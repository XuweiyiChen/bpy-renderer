```python
import objaverse
```

```python
uids = objaverse.load_uids()
```

```python
uids[:1000]
```

```python
# Save the first 1000 UIDs to a text file
uids_1000 = uids[:1000]

# Save to txt file with each UID on a separate row
with open('uids_1000.txt', 'w') as f:
    for uid in uids_1000:
        f.write(uid + '\n')

print(f"Saved {len(uids_1000)} UIDs to 'uids_1000.txt'")
print(f"First 5 UIDs saved:")
for i, uid in enumerate(uids_1000[:5]):
    print(f"  {i+1}. {uid}")
print("...")
print(f"Last 5 UIDs saved:")
for i, uid in enumerate(uids_1000[-5:], len(uids_1000)-4):
    print(f"  {i}. {uid}")
```
